```python
import pandas as pd
```

```python
comuni = "gi_comuni.csv"
provincie = "gi_province.csv"
regioni = "gi_regioni.csv"

comuni = pd.read_csv(comuni, delimiter=";", quotechar='"', dtype=str)
comuni = comuni.applymap(lambda x: x.strip() if isinstance(x, str) else x)[["sigla_provincia", "denominazione_ita_altra"]]
comuni.rename(columns={"denominazione_ita_altra": "Comune"}, inplace=True)

provincie = pd.read_csv(provincie, delimiter=";", quotechar='"', dtype=str)
provincie = provincie.applymap(lambda x: x.strip() if isinstance(x, str) else x)[["sigla_provincia", "denominazione_provincia"]]
provincie.rename(columns={"denominazione_provincia": "Provincia"}, inplace=True)

regioni = pd.read_csv(regioni, delimiter=";", quotechar='"', dtype=str)

display(comuni.head())
print(comuni.shape[0])

display(provincie.head())
print(provincie.shape[0])

display(regioni.head())
print(regioni.shape[0])
```

```python
def imposta_provincia(sigla):
    provincia = provincie.loc[provincie["sigla_provincia"] == sigla, "Provincia"]
    
    # Se la Serie è vuota, restituisci un valore di default (es. "Sconosciuto" o None)
    return str(provincia.iloc[0]) if not provincia.empty else "Sconosciuto"

# Applica la funzione alla colonna "sigla_provincia"
comuni["Provincia"] = comuni["sigla_provincia"].apply(imposta_provincia)
```

```python
comuni.to_csv("Comuni.csv", index=False)
```
